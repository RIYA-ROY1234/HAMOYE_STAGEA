```python
#What is the total number and percentage of missing data in 2014 to 3 decimal places?

missingdata_2014 = df['Y2014'].isna()
missingsum = missingdata_2014.sum()
print(missingsum)
totalrows =len(df['Y2014'])
print(totalrows)
print((missingsum/totalrows)*100)
percentagemissing =(missingsum/totalrows)*100
print(round(percentagemissing,3))
```

```python
#Select columns â€˜Y2017â€™ and â€˜Areaâ€™, Perform a groupby operation on â€˜Areaâ€™.  Which of these Areas had the 7th lowest sum in 2017?
selectedcolumns = df[['Y2017', 'Area']]
grouped= selectedcolumns.groupby('Area')
grouped= selectedcolumns.groupby('Area').sum()
sorted_groups = grouped.sort_values(by='Y2017')
seventh_lowest_area = sorted_groups.index[6]
print(seventh_lowest_area)
```

```python
#Which year had the least correlation with â€˜Element Codeâ€™? 
select_columns = df[['Element Code', 'Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']]
GROUPED = select_columns.groupby('Element Code')
GROUPED = select_columns.groupby('Element Code').corr().unstack()
GROUPED
```

```python
least_correlated_year = GROUPED.min()
least_correlated_year
```

```python
least_correlated_year_overall = least_correlated_year.idxmin()
least_correlated_year_overall
```
